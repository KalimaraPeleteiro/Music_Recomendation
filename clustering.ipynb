```python
import pandas as pd
```

```python
dataset = pd.read_csv("dataset.csv")
dataset.head()
```

##### Preparando o conjunto de dados para o Clustering

```python
# Explicit - Se a música possui letras explícitas
# Key - Qual a chave geral da faixa (teoria musical)
# Mode - Modalidade de uma faixa (teoria musical)

# Retiradas por apresentarem pouca relação direta entre as músicas


dataset.drop(['explicit', 'key', 'mode'], axis = 1)
```

```python
# Criar uma cópia do dataset (alterado), para que possamos salvar o original

from sklearn.preprocessing import OneHotEncoder

# Usando OneHot para não perder informação sobre os artistas
OneHot = OneHotEncoder(dtype=int)
novas_colunas = OneHot.fit_transform(dataset[['artists']]).toarray()

# Criando um dataset com o OneHot
dataset_alterado = dataset.drop('artists', axis = 1)
dataset_alterado = pd.concat([dataset_alterado, pd.DataFrame(novas_colunas, columns=OneHot.get_feature_names_out(['artists']))], axis=1)
```

```python
dataset_alterado.head()
```

```python
# Retirando Colunas Não-Numéricas
dataset_alterado = dataset_alterado.drop(['id', 'name', 'artists_song'], axis=1)
```

```python
dataset_alterado.shape
```

#### Últimos Ajustes (PCA e Padronização)

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

SEED = 10
np.random.seed(SEED)

pipeline = Pipeline([('scaler', StandardScaler()),
                     ('PCA', PCA(n_components = 0.7, random_state=SEED))])
dataset_alterado = pipeline.fit_transform(dataset_alterado)
dataset_alterado = pd.DataFrame(data=dataset_alterado)
```

```python
dataset_alterado.head()
```

#### Clusterizando as Músicas

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=50, verbose=False, random_state=SEED)
model.fit(dataset_alterado)

dataset_alterado['grupo_cluster'] = model.predict(dataset_alterado)
dataset_alterado['artistas'] = dataset['artists']
dataset_alterado['musica'] = dataset['name']
```

```python
dataset_alterado.head()
```

```python
# As colunas 0 e 1 serão usadas como coordenadas espaciais para o recomendador.
dataset_final = dataset_alterado[[0, 1,'grupo_cluster', 'artistas', 'musica']]
dataset_final['id'] = dataset['id']
```

```python
dataset_final
```

```python
dataset_final.to_csv("musicas_agrupadas.csv")
```
